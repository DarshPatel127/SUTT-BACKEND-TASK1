```python
import pandas as pd
import json
```

```python
df = pd.read_csv('Mess Menu Sample - Sheet1.csv')
```

```python
print(df)
```

```python
#Main dictionary for mapping day menu to respective dates
date_mapping = {}

# To get the dates from the first row
dates=df.iloc[0]

# Creating loops for getting dates as keys and also to enumerate the dates as nuremic indices
for idx, date in enumerate(dates):
  # Initialing daily menu
  day_menu={}
  #initiating lists to store values for different meals
  l1=[]
  l2=[]
  l3=[]
  # To get the breakfast items (according to the sheet)
  for i in range (2,11):
    b_val = str(df.iloc[i][idx])
    # To remove all the empty elements and elements which only contain asterisks
    if b_val and all(ch == '*' for ch in b_val) == False:
      l1.append(b_val)
  # Logic applied to lunch and dinner values (identical to the breakfast values code)
  for j in range (13,21):
    l_val = str(df.iloc[j][idx])
    if l_val and all(ch == '*' for ch in l_val) == False:
      l2.append(l_val)
  for k in range (23,29):
    d_val = str(df.iloc[k][idx])
    if d_val and all(ch == '*' for ch in d_val) == False:
      l3.append(d_val)
  # Finally mapping the values of dictionary to their keys
  day_menu['Breakfast']=l1
  day_menu['Lunch']=l2
  day_menu['Dinner']=l3
  date_mapping[date]=day_menu
```

```python
# Serializing to JSON
with open('menu.json', 'w') as f:
  json.dump(date_mapping, f)
```
